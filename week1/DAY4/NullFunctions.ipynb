# LEVEL 1 (BASIC)

```sql
--Show all employees whose salary is NULL
Select * from Employees1 where salary is NULL
```

```sql
--Show all orders where discount is NOT NULL
select * from Orders1 where discount is not null
```

```sql
--Get products where category is NULL
select * from products where category is NULL 
```

```sql
--_ Count number of employees with NULL manager id
select count(*) as count_of_emp from Employees1 where manager_id is NULL
```

# LEVEL 2 (ISNULL)

```sql
--Replace NULL salary with 0
UPDATE Employees1 set salary=0 where salary is null
```

```sql
--Replace NULL bonus with 1000
update Employees1 set bonus = 1000 where bonus is NULL
```

```sql
select * from Employees1
```

```sql
--Show order amount, if NULL replace with 500
select amount, ifnull(amount,500) as order_amount from Orders1
```

```sql
--Replace NULL stock with 0
update products set stock=0 where stock is null
```

# LEVEL 3 (COALESCE)

```sql
--Show employee earnings using: salary, if NULL use bonus
select *,coalesce(salary,bonus) from Employees1
```

```sql
--Show first available value: salary → bonus → 0
SELECT coalesce(salary,bonus,0) as earnings FROM Employees1
     
```

```sql
-- Show product price: price → 1000 (default)
SELECT coalesce(price,1000) as price FROM products
```

```sql
--Get customer payment: amount → discount → 0
select coalesce(amount,discount,0) as payment from Orders1
```

# LEVEL 4 (NULLIF)

```sql
--Convert salary to NULL if salary = 0
select *,NULLIF(salary,0) as updated_salary from Employees1
```

```sql
--Convert discount to NULL if discount = 0
select *,NULLIF(discount,0) as updated_discount from Orders1
```

```sql
--Use NULLIF to avoid divide by zero:
SELECT *,amount / NULLIF(discount, 0)
FROM orders1;
```

```sql
--Replace coupon_code with NULL if it is 'DISC10'
select * , NULLIF(coupon_code,'DISC10')from Orders1
     
```

# LEVEL 5 (REAL-TIME SCENARIOS)

```sql
--Calculate total earnings: salary + bonus (handle NULL properly)
select name,
       coalesce(salary, 0) + coalesce(bonus, 0) as total_earnings
from Employees1;
```

```sql
--Show employees where: both salary AND bonus are NULL
select name
from Employees1
where salary is null 
  and bonus is null;
```

```sql
--Show products where: price is NULL but category is NOT NULL
select product_name
from products
where price is null 
  and category is not null;
```

```sql
--Show orders where: both amount and discount are NULL
select *
from Orders1
where amount is null
  and discount is null;
```

# LEVEL 6 (ADVANCED)

```sql
--Show employee income:COALESCE(salary, bonus, 1000)
select * ,coalesce(salary,bonus,1000)from Employees1
```

```sql
--Replace empty discount with NULL using NULLIF:NULLIF(discount, 0)
update Orders1
set discount=nullif(discount,0)
```

```sql

--Show final payable amount: amount - discount (handle NULL)
select *,coalesce(amount,0) - coalesce(discount,0) from Orders1
```
