```python
df=spark.read.csv("/Volumes/day1/default/exercise2/sample_data.csv",header=True,inferSchema=True)
```

```python
df.display()
```

```python
df=df.replace(["blank","Blank"],None)
```

```python
df.display()
```

```python
from pyspark.sql.functions import *
df=df.withColumn("city",upper(col("city")))
df.display()
```

```python
df=df.withColumn("date",when(col("date").isNull(),None).otherwise(col("date")))
df.display()
```
