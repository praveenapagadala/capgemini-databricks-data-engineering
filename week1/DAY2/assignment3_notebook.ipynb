Basic GROUP BY Queries

```sql
select department, sum(salary) as total_salary from Employee group by department;
```

```sql
select department,count(*) as Number_of_Employees from Employee group by department;
```

```sql
select department, avg(salary) from Employee group by department;
```

```sql
select department, max(salary) from Employee group by department;
```

```sql
select department, min(salary) from Employee group by department;
```

GROUP BY with Conditions

```sql
select department, sum(salary) as total_salary from Employee group by department having total_salary>100000;
```

```sql
select department, count(emp_id) as Num_of_emp from Employee group by department having Num_of_emp>2;
```

```sql
select department,avg(salary) AS avg_salary
from Employee
where "joining_date" > '2021-01-01'
group by department
```

```sql
select department, max(salary) as maximum_salary from Employee group by department
having maximum_salary>75000;
```

```sql
select department, sum(salary) as total_salary from Employee group by department having total_salary<150000;
```

##GROUP BY WITH JOINS


```sql
select e.emp_name,sum(amount) as total_sales from Sales s join Employee e on s.emp_id=e.emp_id group by e.emp_name;
```

```sql
select e.emp_name,count(sales_id) as total_sales from Sales s join Employee e on s.emp_id=e.emp_id group by e.emp_name;
```

```sql
select s.product,sum(amount) as total_sales_amount from Sales s join Employee e on s.emp_id=e.emp_id group by s.product;
```

```sql
select s.product,avg(amount) as avg_sales_amount from Sales s join Employee e on s.emp_id=e.emp_id group by s.product;
```

```sql
select e.emp_name,count(sales_id) as count_of_sales  from Sales s join Employee e on s.emp_id=e.emp_id group by e.emp_name having count_of_sales>2;
```

Advanced GROUP BY

```sql
select e.emp_name,sum(e.salary) as total_salary,count(s.sales_id) as count_of_sales  from Sales s join Employee e on s.emp_id=e.emp_id group by e.emp_name;
```

```sql
select e.emp_name, count(distinct s.product) as unique_product from Sales s join Employee e on s.emp_id=e.emp_id group by e.emp_name;
```

```sql
select e.emp_name,max(amount) as highest_sales_amount from Sales s join Employee e on s.emp_id=e.emp_id group  by e.emp_name ;
```

```sql
select product,sum(amount) as total_sales_amount from Sales group by product having total_sales_amount>50000;
```

```sql
select department, avg(s.amount) as avg_sales
from Sales s
join Employee e on s.emp_id = e.emp_id
group by department order by avg_sales desc limit 1;
```

## Real-World Scenarios

```sql
select department, sum(s.amount) as total_sale
from Sales s
join Employee e on s.emp_id = e.emp_id
group by department order by total_sale desc limit 1;
```

```sql
select e.emp_name, sum(s.amount) as total_sale
from Sales s
join Employee e on s.emp_id = e.emp_id
group by e.emp_name order by total_sale desc limit 3;
```

```sql
select year(e.joining_date),count(e.emp_id) as number_of_emp,avg(e.salary) as avg_salary from Employee e group by year(e.joining_date);
```

```sql
select e.department,sum(amount) as total_sales_amount from Employee e join Sales s on s.emp_id=e.emp_id group by e.department;
```

```sql
select e.emp_name,e.department,count(s.sales_id) as count_of_sales from Employee e join Sales s on e.emp_id=s.emp_id group by e.emp_name,e.department having count_of_sales=0; 
```
