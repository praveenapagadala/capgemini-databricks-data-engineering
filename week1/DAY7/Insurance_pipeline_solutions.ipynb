```python
premium_df = policies_clean.groupBy("customer_id") \
    .agg(sum("premium").alias("total_premium"))
```

```python
claims_df = claims_clean.groupBy("policy_id") \
    .agg(sum("claim_amount").alias("total_claim"))

claims_customer = claims_df.join(policies_clean, "policy_id") \
    .groupBy("customer_id") \
    .agg(sum("total_claim").alias("total_claim"))
```

```python
final_df = premium_df.join(claims_customer, "customer_id", "left")
```

```python
final_df = final_df.withColumn(
    "risk_score",
    col("total_claim") / col("total_premium")
)
```

```python
final_df = final_df.fillna({"total_claim": 0})
```

```python
risk_df = final_df.withColumn(
    "risk_score",
    col("total_claim") / col("total_premium")
)
```

#### creating temp views

```python
premium_df.createOrReplaceTempView("premium")
claims_customer.createOrReplaceTempView("claims")
risk_df.createOrReplaceTempView("risk")
customers.createOrReplaceTempView("customers")
policies_clean.createOrReplaceTempView("policies")
claims_clean.createOrReplaceTempView("claims")
policy_agent.createOrReplaceTempView("policy_agent")
agents.createOrReplaceTempView("agents")
```

#### using cte

```python
%sql
with combined as (
    select p.customer_id,
           p.total_premium,
           c.total_claim
    from premium p
    left join claims c
    on p.customer_id = c.customer_id
),
risk as (
    select *,
           total_claim / total_premium as risk_score
    from combined
)
select * from risk;
```

```python
customers_clean.createOrReplaceTempView("customers")
```

#### top 3 risky customers per city

```python
%sql
select *
from (
    select c.city, r.customer_id, r.risk_score, row_number() over (partition by c.city order by r.risk_score desc) as rn
    from risk r
    join customers c on r.customer_id = c.customer_id
) t
where rn <= 3;
```

#### rank agents by premium

```python
%sql
select agent_id, sum(premium) as total_premium, rank() over (order by sum(premium) desc) as rank
from policies p
join policy_agent pa on p.policy_id = pa.policy_id
group by agent_id;
```
