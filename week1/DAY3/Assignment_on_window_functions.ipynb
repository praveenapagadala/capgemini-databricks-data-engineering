## ROW_NUMBER()

```sql
--1
select emp_name,emp_id,salary,row_number() over( order by salary desc) as Row_number from employees_ ;
```

```sql
--2
select emp_name,emp_id,salary,department,row_number() over( partition by department order by salary desc) as Row_number from employees_ ;
```

```sql
--3
select emp_name,emp_id,join_date,row_number() over( partition by join_date order by join_date desc) as Row_number from employees_ ;
```

```sql
--4
select emp_name,emp_id,join_date,department,row_number() over( partition by department order by join_date desc) as Row_number from employees_ ;
```

# USING RANK()

```sql
--9
select emp_id,emp_name ,salary,rank() over (order by salary desc) as ranking from employees_;
```

```sql
--10
select emp_id,emp_name ,department,salary,rank() over (partition by department order by salary) as ranking from employees_;
```

```sql
--11
select emp_id,emp_name ,join_date,rank() over (order by join_date desc) as ranking from employees_;
```
