Problem 1: Salary Hike Based on Experience and Performance 

```sql
select emp_name,salary,experience,performance_rating,
case 
when experience>=8 and performance_rating='A' then salary*1.2
when experience>=5 and performance_rating='B' then salary*1.15
when performance_rating='C' then 0
else salary*1.1
end as salary_hike
from EMP;
```

2. Problem 2: Bonus Calculation Based on Department and Performance

```sql
select department,performance_rating,
case 
when department='Finance' and performance_rating ='A' then salary*1.2
when department='Finance' and performance_rating ='B' then salary*1.15
when department='Finance' and performance_rating ='C' then salary*0.05 
when department='Engineering' and performance_rating ='A' then salary*1.18
when department='Engineering' and performance_rating ='B' then salary*1.12
when department='Engineering' and performance_rating ='C' then salary*0.03
else salary*1.1
end as Bonus
from EMP
```

using nested case

```sql
select department, performance_rating,
case 
    when department = 'Finance' then
        case 
            when performance_rating = 'A' then salary * 1.2
            when performance_rating = 'B' then salary * 1.15
            when performance_rating = 'C' then salary * 0.05
        end

    when department = 'Engineering' then
        case 
            when performance_rating = 'A' then salary * 1.18
            when performance_rating = 'B' then salary * 1.12
            when performance_rating = 'C' then salary * 0.03
        end

    else salary * 1.1
end as Bonus
from EMP;
```

## Nested case and when 

Problem 1: Nested CASE for Performance and Salary Hike Based on Multiple Criteria 

```sql
select performance_rating,salary,experience,
case 
  when performance_rating='A' then 
    case 
      when salary>80000 and experience>5 then salary*1.25
      when salary between 50000 and 80000 then salary*1.15
      else salary*1.20
    end  
  when performance_rating='B' then 
    case 
      when experience>5 then salary*1.12
      else salary*1.10
    end
  when performance_rating='C' then salary
end as salary_hike
from EMP;
```

2.Department and Performance 

Scenario: You are tasked with giving bonuses based on the department, performance rating, and experience: 

```sql
select department,performance_rating,experience,salary,
case 
  when department='Finance' then 
    case 
      when  performance_rating='A' and experience>10 then salary*1.25
      else salary*1.20
    end 
  when department='HR' then 
    case 
      when performance_rating='B' and experience>5 then salary*1.15
      else salary*1.10
    end
  else salary*1.10
end as bonus
from EMP;
```

## Problem 3: Nested CASE for Employee Categorization

 Based on Salary, Performance, and ExperienceScenario: Categorize employees based on their salary, performance rating, and experience: 

```sql
select salary,performance_rating,experience,
case 
  when salary>70000 then 
    case 
      when performance_rating='A' and experience>8 then 'Top Performer'
      when experience<8 then 'Mid Performer'
    end
  when salary between 50000 and 70000 then 
    case 
      when performance_rating='A'then 'Rising Star'
      else 'Average Performer'
    end 
  when salary<50000 then 'Low Performer'
end as Label
from EMP;
```

## 4.Nested CASE for Tax Bracket Based on Salary and Experience 

Scenario: You want to determine the tax bracket of employees based on their salary and experience: 

```sql
select *,
case 
when salary>=90000 then 
  case 
    when experience>=10 then '35%'
    else '30%'
  end
when salary between 60000 and 90000 then 
  case 
    when experience>=5 then '25%'
    else '5%'
  end
when salary<=60000 then '15%'
end as Tax_Rate
from EMP;
```

## Problem 5: Nested CASE for Promotion Eligibility Based on Performance, Salary, and Experience 

Scenario: You want to determine if employees are eligible for promotion based on their performance rating, salary, and experience: 

```sql
select salary,performance_rating,experience,
case 
  when performance_rating='A' then 
    case 
      when salary>75000 and experience>7 then 'Eligible for Senior Role'
      else 'Eligible for junior Role'
    end 
  when performance_rating='B' then 
    case 
      when experience>5 then 'Eligible for Consideration'
      else 'Not Eligible'
    end
  else 'Not Eligible'
  end as promotion_eligibility
  from EMP;
```
