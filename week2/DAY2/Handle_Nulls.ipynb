### create dataset with some nulls

```python

data = [
    (1, "A", None),
    (2, "B", 5000),
    (3, None, 7000),
    (4, "D", None)
]
df = spark.createDataFrame(data, ["id", "name", "salary"])
df1=df = spark.createDataFrame(data, ["id", "name", "salary"])
```

```python

df.display()
df.printSchema()
```

### isNull() / isNotNull()

```python
df.filter(df.salary.isNull()).display()
```

### fillna()

```python
df.fillna({"salary": 0}).display()
```

### dropna()

```python
df.dropna().display()
```

### coalesce()

```python
from pyspark.sql.functions import coalesce, lit

df.withColumn("salary_new", coalesce(df.salary, lit(0))).display()
```

###  Task


- Replace NULL salary with avg salary
- Drop rows where name is NULL
- Count NULL records

```python
from pyspark.sql.functions import avg, col, when

avg_salary = df.select(avg(col("salary"))).first()[0]
df.withColumn("salary", when(col("salary").isNull(), avg_salary).otherwise(col("salary"))).display()
```

```python
df.dropna(subset=["name"]).show()
```

```python
df1.filter(col("id").isNull() | col("name").isNull() | col("salary").isNull()).count()
```

### Type Casting + Column Operations

### Casting

```python
df = df.withColumn("salary", df.salary.cast("int"))
df.printSchema()
```

### Column Operations

```python
from pyspark.sql.functions import col, when

df.select("id", "salary").display()

df.withColumn("bonus", col("salary") * 0.1).display()

df.withColumn("category",
    when(col("salary") > 6000, "High")
    .otherwise("Low")
).display()
```

### Replace the above with below

```python
# Step 1: Replace NULL salary with 1000
df = df.withColumn(
    "salary",
    when(col("salary").isNull(), 1000)
    .otherwise(col("salary"))
)

# Step 2: Add bonus column (10% of salary)
df = df.withColumn(
    "bonus",
    col("salary") * 0.1
)

# Step 3: Add category column
df = df.withColumn(
    "category",
    when(col("salary") > 6000, "High")
    .otherwise("Low")
)

# Final output
print("Final Data:")
df.display()
```

```python
def grade(salary):
    if salary is None:
        return "Unknown"
    elif salary > 6000:
        return "A"
    else:
        return "B"
```

```python
from pyspark.sql.functions import udf
from pyspark.sql.types import StringType

grade_udf = udf(grade, StringType())
```

```python
df = df.withColumn("grade", grade_udf(df.salary))
df.display()
```

### Same logic WITHOUT UDF:

```python
from pyspark.sql.functions import when, col

df = df.withColumn(
    "grade",
    when(col("salary").isNull(), "Unknown")
    .when(col("salary") > 6000, "A")
    .otherwise("B")
)
```

### This is
- Faster
- optimized
- Recommended

### WHEN SHOULD YOU USE UDF?
- Only when: Complex logic
- No built-in alternative
- Custom transformations
