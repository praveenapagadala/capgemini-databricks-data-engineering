```sql
CREATE TABLE Employeess (
    emp_id INT PRIMARY KEY,
    name VARCHAR(50),
    department_id INT,
    salary INT,
    manager_id INT
);
```

```sql
drop table if exists Departments;

create table Departments (
    dept_id int primary key,
    dept_name varchar(50)
);
```

```sql

drop table if exists Orders;
CREATE TABLE Orders (
    order_id INT PRIMARY KEY,
    emp_id INT,
    order_amount INT,
    order_date DATE
);
```

# Using Subqueries, Exists, All operators

```sql
--1
select emp_id,salary from Employeess where salary>(select avg(salary) as avg_sal from Employeess);
```

```sql
--2
select e.*, d.dept_name from Employeess e join Departments d on e.department_id = d.dept_id
where e.department_id = (select department_id from Employeess where name = 'Alice'
);
```

```sql
--3
select * from Employeess where salary=(select min(salary) from Employeess)
```

```sql
--4
select e.* from Employeess e where exists (select 1 from Orders o where e.emp_id = o.emp_id)
```

```sql
--5
select *
from Employeess
where salary > (select max(salary) from Employeess where department_id = 102);
```

# USING CTEs

```sql
--1
with highest_sal as 
(select * from Employeess where salary>50000)
select * from highest_sal; 
```

```sql
--2
with dept_avg as(select avg(salary)as avg_salary from Employeess group by department_id)
select * from dept_avg;
```

```sql
--3
with emp_with_dept as
(select d.dept_id,e.name from Departments d join Employeess e on d.dept_id=e.department_id)
select * from emp_with_dept;
```

```sql
--4
with emp_orders as
(select e.emp_id,sum(order_amount) as total_amount from Employeess e join Orders o on e.emp_id=o.emp_id group by e.emp_id)
select * from emp_orders;
```

```sql
--5
with dept_avg as (
    select department_id, avg(salary) as avg_salary
    from Employeess
    group by department_id
)
select e.emp_id, e.name, e.salary, d.dept_name
from Employeess e
join dept_avg d
on e.department_id = d.department_id
join Departments d
on e.department_id = d.dept_id
where e.salary > d.avg_salary;
```
